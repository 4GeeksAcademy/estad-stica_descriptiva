### Exercise 1

We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
# TODO
import numpy as np
import statistics as stats
from scipy.stats import skew 
from scipy.stats import kurtosis

#Generate an array of 100 elements following a normal distribution.

mu, sigma = 1, 2 # mean and standard deviation as required by the random.normal documentation those are required parameters
Array_normal = np.random.normal(mu, sigma, 100)

# Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
Array_chi=np.random.chisquare(3, size=100)

#Calculate the main metrics and statistical measures that best describe the two vectors.

print("Main Metrics of the array with normal distribution")
print("-----------------------------------------")
Mean1= stats.mean(Array_normal)
print("La media del array es ", Mean1)
Median1=stats.median(Array_normal)
print("La mediana del array es ", Median1)
Mode1=stats.mode(Array_normal)
print("La moda del array es ", Mode1)

print("\nMeasures of dispersion")
print("-----------------------------------------")

range_ = max(Array_normal) - min(Array_normal)
print(f"El rango es de: {range_}")

variance = stats.variance(Array_normal)
std = stats.stdev(Array_normal)
print(f"Variance: {variance}")
print(f"Standard deviation: {std}")

print("\nPosition measures")
print("-----------------------------------------")
percentile_value = np.percentile(Array_normal, 100)
quantile_value = np.quantile(Array_normal, 1)

print("El percentil tomando como valor 100 es: ", percentile_value)
print("El cuantil tomando como valor 1 es: ", quantile_value)


print("\n Shape measures")
print("-----------------------------------------")
skewness=skew(Array_normal)
print("La asimetría del Array con distribucion normal es de: ", skewness)
kurt=kurtosis(Array_normal)
print("La kurtosis del Array con distribucion normal es de: ", kurt, "\n")

print("*****************************************")
print("-----------------------------------------")
print("*****************************************")


print("Main Metrics of the array with chi square distribution")
print("-----------------------------------------")
Mean1= stats.mean(Array_chi)
print("La media del array es ", Mean1)
Median1=stats.median(Array_chi)
print("La mediana del array es ", Median1)
Mode1=stats.mode(Array_chi)
print("La moda del array es ", Mode1)

print("\nMeasures of dispersion")
print("-----------------------------------------")

range_ = max(Array_chi) - min(Array_chi)
print(f"El rango es de: {range_}")

variance = stats.variance(Array_chi)
std = stats.stdev(Array_chi)
print(f"Variance: {variance}")
print(f"Standard deviation: {std}")

print("\nPosition measures")
print("-----------------------------------------")
percentile_value = np.percentile(Array_chi, 100)
quantile_value = np.quantile(Array_chi, 1)

print("El percentil tomando como valor 100 es: ", percentile_value)
print("El cuantil tomando como valor 1 es: ", quantile_value)


print("Shape measures")
print("-----------------------------------------")
skewness=skew(Array_chi)
print("La asimetría del Array con distribucion normal es de: ", skewness)
kurt=kurtosis(Array_chi)
print("La kurtosis del Array con distribucion normal es de: ", kurt)



````

Write a Python program to calculate the standard deviation of the following data:

````python
# TODO
data=[4,2,5,8,6]
st_dev=np.std(data)
print("La desviación estandar es de:",st_dev)

#Sin usar numpy

def calculate_standard_deviation (data):

# Media es igual a la suma de todos sus datos dividido entre el numero total de datos
    mean = sum(data) / len(data)
# Variance es igual al cuadrado de las diferencias
    variance = sum((x - mean) ** 2 for x in data) / len(data)
# Raíz cuadrada de la varianza para obtener desviación estándar
    st_dev = variance ** 0.5

print("La desviación estándar es de:", st_dev)
````
